# Small example demonstrating sampling of noise

```python
import torch
import matplotlib.pyplot as plt
from trawl_sinn import GaussianNoise, ExponentialNoise
```

```python
noise = GaussianNoise(mean = 1.0, var = 0.5)
times = torch.linspace(0.0, 1.0, 100)
```

```python
data = noise.sample(times, batch_size = 1000, unsqueeze_last = False, batch_first = True)
plt.figure(figsize=(8, 4))
plt.plot(times.numpy(), data.numpy().T, color = 'blue', alpha = 0.005)
plt.title('Density of the trajectories')
plt.xlabel('time')
plt.ylabel('White Noise')
plt.grid(True)
plt.show()
```

```python
noise = ExponentialNoise(rate = 0.5)
times = torch.linspace(0.0, 1.0, 100)
```

```python
data = noise.sample(times, batch_size = 1000, unsqueeze_last = False, batch_first = True)
plt.figure(figsize=(8, 4))
plt.plot(times.numpy(), data.numpy().T, color = 'blue', alpha = 0.005)
plt.title('Density of the trajectories')
plt.xlabel('time')
plt.ylabel('Exponential Noise')
plt.grid(True)
plt.show()
```
